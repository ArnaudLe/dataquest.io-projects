# Introduction

```python
import pandas as pd

# Read csv file
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

# Some genreal info
cols = star_wars.columns
print(star_wars.shape)

# Remove rows where RespondentID == NaN
star_wars = star_wars[star_wars["RespondentID"].notnull()]
print(star_wars.shape)

star_wars
```

# Cleaning Up Yes/No Columns

```python
# Series containing Yes/No
seen = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"]
fan = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]

# Dictionary
yes_no = {
    "Yes": True,
    "No": False
}

# Mapping
seen = seen.map(yes_no)
fan = fan.map(yes_no)

# Convert
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = seen
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = fan

star_wars.head(10)
```

# Cleaning Up Checkbox Columns

```python
import numpy as np

# Remapping dictionary
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

# Convert to Boolean
for c in star_wars.columns[3:9]:
    star_wars[c] = star_wars[c].map(movie_mapping)
    
# Rename cols
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

# Check
star_wars.head(10)
```

# Cleaning Up Ranking Columns

```python
# Convert to float
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

# Rename cols
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

# Check
star_wars.head(10)
```

# Finding The Highest Ranked Movie

```python
# Compute the mean (rank) for each movie
means = star_wars[star_wars.columns[9:15]].mean()

# Bar chart
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(range(6), means)
```

Empire Strikes Back is the best Star Wars movie!

# Finding The Most Seen Movie

```python
# Sum on seen movies
seen = star_wars[star_wars.columns[3:9]].sum()

# Bar chart
plt.bar(range(6), seen)
plt.xlabel("Star Wars movie")
plt.ylabel("Number of times")
plt.title("How many times Star Wars was seen")
```

A lot of people saw Star Wars V :o

# Exploring The Data By Binary Segments

```python
# Split data into 2 subsets
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

# Charts
fig = plt.figure(figsize=(10,10))

# Males
mean_males = males[males.columns[9:15]].mean()
sum_males = males[males.columns[3:9]].sum()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax1.bar(range(6), mean_males)
ax2.bar(range(6), sum_males)

ax1.set_title("Mean ranking - Male")
ax2.set_title("Sum views - Male")

# Females
mean_females = females[females.columns[9:15]].mean()
sum_females = females[females.columns[3:9]].sum()

ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax3.bar(range(6), mean_females)
ax4.bar(range(6), sum_females)

ax3.set_title("Mean ranking - Female")
ax4.set_title("Sum views - Female")
```
